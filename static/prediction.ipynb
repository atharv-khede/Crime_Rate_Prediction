# MURDER PREDICTION

```python
x = file.loc[:, ['STATE/UT', 'DISTRICT', 'YEAR']]
y = file['MURDER']
```

```python
# from sklearn.preprocessing import LabelEncoder

# le = LabelEncoder()
# x['STATE/UT'] = le.fit_transform(x['STATE/UT'])
# x['DISTRICT'] = le.fit_transform(x['DISTRICT'])
# x
```

```python
catagorical = x.select_dtypes(include=['object'])
catagorical.head()
```

```python
data_dummy = pd.get_dummies(catagorical, drop_first=True)
data_dummy
```

```python
x = x.drop(list(catagorical.columns), axis=1)
x.head()
```

```python
x = pd.concat([x, data_dummy], axis=1)
x
```

```python
from sklearn.preprocessing import scale

col = x.columns   # store columns before scalling 
x = pd.DataFrame(scale(x))
x.columns = col
x.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=100)
X_train
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

lm.fit(X_train, y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
X_test.shape
```

```python
from sklearn.metrics import r2_score
y_pred = lm.predict(X_test)

print(r2_score(y_true=y_test, y_pred=y_pred))
```
