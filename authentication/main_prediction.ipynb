```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
file1 = pd.read_csv('./static/Dataset/01_District_wise_crimes_committed_IPC_2001_2012.csv')
file2 = pd.read_csv('./static/Dataset/01_District_wise_crimes_committed_IPC_2013.csv')
```

```python
frames = [file1, file2] 
file= pd.concat(frames)
file.shape
```

```python
file.columns
```

```python
file1 = file[['MURDER', 'RAPE', 'KIDNAPPING & ABDUCTION','DACOITY','ROBBERY', 'BURGLARY','THEFT', 'DOWRY DEATHS']]
```

```python
filt = file[file['DISTRICT']=='TOTAL'].index
file.drop(filt, inplace=True)
file
filt = file[file['DISTRICT']=='ZZ TOTAL'].index
file.drop(filt, inplace=True)
file
```

```python
file['STATE/UT'] = file['STATE/UT'].str.upper()
file['DISTRICT'] = file['DISTRICT'].str.upper()
```

```python
file1 = file[['MURDER', 'RAPE', 'KIDNAPPING & ABDUCTION','DACOITY','ROBBERY', 'BURGLARY','THEFT', 'DOWRY DEATHS']]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)

kmeans.fit(file1)

labels = kmeans.labels_

centroids = kmeans.cluster_centers_

score = kmeans.score(file1)
ymeans = kmeans.fit_predict(file1)
ymeans
```

```python
file1['label']= ymeans
```
